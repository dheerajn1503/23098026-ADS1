```python
#function to plot bar graph

def plot_spec_distribution():
    spec_count = df['Species'].value_counts()
    spec_count.plot(kind='bar', color=['blue', 'orange', 'green'])
    plt.title('Species Distribution (Bar Graph)')
    plt.xlabel('Species')
    plt.ylabel('Count')
    plt.show()

# Call the function to display the plot
plot_spec_distribution()
```

```python
#function to plot scatterplot of two features

def plot_sepal_petal_scatter():
    sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', data=df, hue='Species')
    plt.title('Sepal Length vs Petal Length (Scatter Plot)')
    plt.xlabel('Sepal Length in cm')
    plt.ylabel('Petal Length in cm')
    plt.show()

# Call the function to display the plot
plot_sepal_petal_scatter()
```

```python
# Compute the correlation matrix for numerical columns only
correlation_matrix = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].corr()

def plot_corr_heatmap():
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()

# Call the function to display the heatmap
plot_corr_heatmap()
```

```python
#Histogram

plt.figure(figsize=(11, 6))
plt.hist(df['SepalLengthCm'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Sepal Length vs Petal Length (Histogram)')
plt.xlabel('SepalLenthCm')
plt.ylabel('PetalLengthCm')
plt.show()
```
